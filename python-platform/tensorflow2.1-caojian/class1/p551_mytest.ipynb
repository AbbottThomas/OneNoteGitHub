```python
# 用梯度法,求函数的最终
# 设置学习率,迭代次数,初始x
lr ,epoch  = 0.1, 100 
# 要加上float32,否出错
x0 = tf.constant(0,dtype=tf.float32)

# lr = tf.convert_to_tensor(lr)
x0 = tf.Variable(tf.constant(0,dtype=tf.float32))
for i in range(epoch):
    with tf.GradientTape() as tape:
        y = tf.square(x0 - 1)
    grads = tape.gradient(y,x0)
    # y=tf.square(x0 - 1)
    # 出错了,直接相上面那样调用
    # grads = tf.GradientTape().gradient(y,x0)

    bb = grads*lr
    # x0 = x0 - bb
    # x0.assign_sub(lr*grads)
    print(i,"次迭代","梯度 ",grads.numpy()," 移动 :",bb.numpy()," 位置",x0.numpy())
```
